```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img_height, img_width = 224, 224
batch_size = 32
num_classes = 488  
```

```python
name_model = 'MobileNetV2'

if name_model == 'EfficientNetB0':
    model = load_model('../models/EfficientNetb0.h5', compile=True)
    feature_layer_name = 'top_conv'
elif name_model == 'MobileNetV2':
    model = load_model('../models/MobileNetV2.h5', compile=True)
    feature_layer_name = 'Conv_1'
elif name_model == 'ResNet50':
    model = load_model('../models/ResNet50.h5' , compile=True)
    feature_layer_name = 'conv5_block3_out'
```

```python
feature_extractor = tf.keras.Model(inputs=model.input, outputs=model.get_layer(feature_layer_name).output)
```

```python
datagen = ImageDataGenerator(rescale=1.0 / 255.0)

train_generator = datagen.flow_from_directory(
    "../data/train",
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False 
)

val_generator = datagen.flow_from_directory(
    "../data/val",
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False 
)

test_generator = datagen.flow_from_directory(
    "../data/test",
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False 
)
```

```python
def extract_features(data_generator, feature_extractor, file="../feature_maps/train_feature_maps.npy", labels_file="../feature_maps/train_labels.npy"):
    feature_maps_list = []
    labels_list = []

    for i in range(len(data_generator)):
        images, batch_labels = data_generator[i]
        feature_maps = feature_extractor.predict(images)
        feature_maps_list.append(feature_maps)
        labels_list.append(batch_labels)  

        if i == len(data_generator) - 1:
            break 

    feature_maps_array = np.concatenate(feature_maps_list, axis=0)
    labels_array = np.concatenate(labels_list, axis=0)

    file_dir = os.path.dirname(file)
    if not os.path.exists(file_dir):
        os.makedirs(file_dir)
    
    labels_file_dir = os.path.dirname(labels_file)
    if not os.path.exists(labels_file_dir):
        os.makedirs(labels_file_dir)

    np.save(file, feature_maps_array)
    np.save(labels_file, labels_array)
```

```python
extract_features(train_generator, feature_extractor, file=f"../feature_maps/{name_model}/train_feature_maps.npy", labels_file=f"../feature_maps/{name_model}/train_labels.npy")
```

```python
extract_features(val_generator, feature_extractor, file=f"../feature_maps/{name_model}/val_feature_maps.npy", labels_file=f"../feature_maps/{name_model}/val_labels.npy")
```

```python
extract_features(test_generator, feature_extractor, file=f"../feature_maps/{name_model}/test_feature_maps.npy", labels_file=f"../feature_maps/{name_model}/test_labels.npy")
```

```python
import json

class_indices = train_generator.class_indices 

with open(f"../feature_maps/{name_model}/class_indices.json", "w") as f:
    json.dump(class_indices, f)
```
